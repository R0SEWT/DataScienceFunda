```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.jointplot(y='age',x='fare',data=titanic,kind='scatter')
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.boxplot(x='class',y='age',data=titanic, palette='rainbow')
#sns.boxplot(x='class',y='age',data=titanic, hue='class', palette='rainbow', legend=False) # current command
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.swarmplot(x='class',y='age',data=titanic, palette='Set2')
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.catplot(x='sex',y='age' ,data=titanic, kind='bar') # perdon 
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.heatmap(titanic.corr(), cmap='coolwarm')
#sns.heatmap(titanic.corr(),cmap='coolwarm', annot=False) # current command
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
g = sns.FacetGrid(data=titanic,col='sex',row='age')
```
