## Datos de entrenamiento y prueba

Ahora que hemos explorado un poco los datos, sigamos adelante y dividamos los datos en conjuntos de entrenamiento y prueba.
**Establezca una variable 'X' igual a las características numéricas de los clientes y una variable 'y' igual a la columna "Monto gastado anual" (Yearly Amount Spent).**

```python
# aaaa lei las instrucciones al reves xd
y = clientes['Yearly Amount Spent']
X = clientes[['Avg. Session Length', 'Time on App', 'Time on Website','Length of Membership']]
#X = clientes.select_dtypes(include=[np.number])
```

```python
from sklearn.model_selection import train_test_split
```

**Utiliza model_selection.train_test_split de sklearn para dividir los datos en conjuntos de entrenamiento y prueba. Establece test_size = 0.3 y random_state = 101**

```python

x_train , x_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

## Entrenando el Modelo

¡Ahora es el momento de entrenar el modelo con nuestros datos de entrenamiento!

**Importar LinearRegression from sklearn.linear_model**

```python
from sklearn.linear_model import LinearRegression
```

**Crear una instancia de un modelo de LinearRegression() llamado lm.**

```python
lm = LinearRegression()
```

**Entrenar/ajustar (Train/fit) el modelo de regresion lineal 'lm' con los datos de entrenamiento.**

```python
lm.fit(x_train, y_train) 
```

**Imprime los coeficientes del modelo**

```python
lm.coef_ 
```

## Predicción de datos de prueba
Ahora que hemos ajustado nuestro modelo, ¡evaluemos su rendimiento prediciendo los valores de prueba!

**Utilice lm.predict() para predecir el conjunto de datos X_test.**

```python
prediccion = lm.predict(x_test)
```

**Crear una gráfica de dispersión de los valores de prueba reales frente a los valores predichos.**

## Evaluando el Modelo

Evaluemos el rendimiento de nuestro modelo calculando la suma residual de cuadrados y la puntuación de varianza explicada (R ^ 2).

**Calcule el error medio absoluto, el error cuadrático medio y la raíz del error cuadrático medio. Consulte la conferencia o Wikipedia para conocer las fórmulas.**

```python
# calcular estas métricas a mano!
```

```python

# calcular estas métricas a mano! 
len(prediccion) # son 150 puntos :c, lo hare con aritmetica c:

# EAM (Error medio absoluto)
# promedio de errores (distancias)
# EAM =  sum |y-y'| / n
EAM = sum(abs(y_test - prediccion)) / len(prediccion)

# MSE (Error cuadratico medio)
# promedio de errores al cuadrado (Penaliza mas los errores grandes)
# MSE = sum (y-y')^2 / n
MSE = sum((y_test - prediccion)**2) / len(prediccion)

# RMSE (Raiz del error cuadratico medio)
# Raiz cuadrada del MSE, regresa a la escala origen
RMSE = np.sqrt(MSE) 

print (f'EAM: {EAM}') # 7.2 dolares de error promedio
print (f'MSE: {MSE}') # -
print (f'RMSE: {RMSE}')  # 8.9 dolares de error promedio, ta bien


```

### Informacion: App supremacy

El tiempo gastado en la app impacta en 20 167% mas en los dolares gastados por el usuario que el tiempo gastado en website.
Esto significa que el 1% del tiempo en la app reporta mayores ingresos que si se duplicara el tiempo en el website. (0.385 > 0.380, respectivamente)

### Conclusion: Mejorar la app y vigilar el website

En primer orden, el impacto del tiempo en la app en los ingresos valida postivamente el modelo de negocio ecommerce - experiencial que maneja la empresa y le aporta valor.

En segundo orden de importancia, se debe clarificar la razon por la que la web esta influyendo tan poco ingresos tan bajos.

Es factible tambien, que el website sea 'doloroso' para el usuario o sea muy diferente de la app.

```python
print('El impacto del tiempo en el website es de 0.19 dolares y el impacto del tiempo en la app es de 38.59 dolares')
print(f'El impacto del tiempo en la app es {round(lm.coef_[1]/lm.coef_[2]*100, 2)}% a comparacion de en el website') 

c_app = lm.coef_[1]/ 100 # 1% de tiempo en la app
c_web = lm.coef_[2] * 2 # 200% de tiempo en el website

print(f'Por cada 1% de tiempo en la app, se gasta {c_app} dolares mas')
print(f'Por cada 200% de tiempo en el website, se gasta {c_web} dolares mas')
```
